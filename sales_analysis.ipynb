```python
# groupings on a categorical column and compute the mean of a numerical column for each group
grouped_means = df.groupby('Product')[['Quantity Sold', 'Revenue ($)']].mean()
print(grouped_means)
```

```python
# patterns or interesting findings from my analysis
# From the grouped analysis, I found that although the Mouse is sold the most (11.5 units on average), it contributes the least to revenue, suggesting it's a high-demand but low-value item. In contrast, Laptops are sold less frequently (4.6 units) but generate the highest revenue ($4600), highlighting their value as premium products. Products like Keyboards and Headphones strike a balance between quantity and revenue, making them consistent mid-level performers.
```

```python
# Task 3: Data Visualization

# line chart - showing sales trend over time
# grouping by date to get total revenue per day
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_data.csv')

daily_sales = df.groupby('Date')['Revenue ($)'].sum()

plt.figure(figsize=(10, 5))
plt.plot(daily_sales.index, daily_sales.values, marker='x')
plt.title('Revenue trend over time')
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# bar chart - average revenue per product
avg_revenue = df.groupby('Product')['Revenue ($)'].mean()

plt.figure(figsize=(8, 5))
avg_revenue.plot(kind='bar', color='pink')
plt.title('Average revenue per product')
plt.xlabel('Product')
plt.ylabel('Average revenue ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# histogram — distribution of quantity sold
plt.figure(figsize=(8, 5))
plt.hist(df['Quantity Sold'], bins=10, color='pink', edgecolor='black')
plt.title('Distribution of Quantity Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# scatter plot — quantity sold vs. revenue
plt.figure(figsize=(8, 5))
plt.scatter(df['Quantity Sold'], df['Revenue ($)'], color='pink', edgecolor='black')
plt.title('Quantity Sold vs. Revenue')
plt.xlabel('Quantity Sold')
plt.ylabel('Revenue ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
